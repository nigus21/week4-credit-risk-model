```python
# ==========================================
# Exploratory Data Analysis (EDA) Notebook
# Credit Risk Modeling Project - Week 4
# ==========================================

import sys
import os

# Add project root to Python path
sys.path.append(os.path.abspath(os.path.join('..')))


# Import modules
import pandas as pd
from src import data_processing as dp
```

```python
# ------------------------------------------
# 1. Load Data
# ------------------------------------------
file_path = "../data/raw/data.csv"  # Update path if needed
df = dp.load_data(file_path)
```

```python
# ------------------------------------------
# 2. Validate Required Columns
# ------------------------------------------
required_cols = ['TransactionId','CustomerId','Amount','Value','ProductCategory',
                 'ChannelId','ProviderId','CountryCode','PricingStrategy','TransactionStartTime']
dp.validate_columns(df, required_cols)
```

```python
# ------------------------------------------
# 3. Data Overview
# ------------------------------------------
overview = dp.data_overview(df)
print("Data Overview:")
for k,v in overview.items():
    print(f"{k}: {v}")
```

```python

# ------------------------------------------
# 4. Summary Statistics
# ------------------------------------------
print("\nNumerical Summary Statistics:")
num_stats = dp.summary_statistics(df)
display(num_stats)
# Missing Values
print("\nMissing Values:")
missing_df = dp.missing_values(df)
display(missing_df)
```

```python
# ------------------------------------------
# 5. Numerical Feature Distributions
# ------------------------------------------
numerical_cols = ['Amount','Value']
dp.plot_numerical_distribution(df, numerical_cols)
```

```python
# ------------------------------------------
# 6. Categorical Feature Distributions
# ------------------------------------------
categorical_cols = ['ProductCategory','ChannelId','ProviderId','CountryCode','PricingStrategy']
dp.plot_categorical_distribution(df, categorical_cols)

# Display top categories for quick insights
for col in categorical_cols:
    print(f"\nTop categories in {col}:")
    print(df[col].value_counts().head(10))
```

```python
# ------------------------------------------
# 7. Correlation Analysis
# ------------------------------------------
dp.correlation_matrix(df, numerical_cols)
```

```python
# ------------------------------------------
# 8. Aggregation Example (Customer-level)
# ------------------------------------------
agg_df = dp.aggregate_transactions(df)
print("\nAggregated Customer-level Features:")
display(agg_df.head())
```
